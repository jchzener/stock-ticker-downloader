```python
import requests
import os
import json
import random
```

```python
session = requests.Session()

session.headers.update({
            'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:85.0) Gecko/20100101 Firefox/85.0'
        })

session.headers

ExchangeConfig_name = "nasdaq" 
ExchangeConfig_api_params = {"exchange": "nasdaq", "offset": "0"}

params = {"tableonly": "true", "download": "true", **ExchangeConfig_api_params}
params
```

```python
# Make API request
response = session.get(
    "https://api.nasdaq.com/api/screener/stocks", params=params, timeout=30
)
```

```python
data = response.json()
ticker_count = len(data.get('data', {}).get('rows', []))
ticker_count
```
