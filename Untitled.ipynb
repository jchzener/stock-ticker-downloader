```python
session = requests.Session()

session.headers.update({
            'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:85.0) Gecko/20100101 Firefox/85.0'
        })

session.headers

api_params = {"exchange": "nyse", "offset": "0"}
api_params_no = {"exchange": "nyse"}


params = {"tableonly": "true", "download": "true", **api_params}
params_no = {"tableonly": "true", "download": "true", **api_params_no}
params
```

```python
# Make API request
response = session.get("https://api.nasdaq.com/api/screener/stocks", params=params, timeout=30)
response_no = session.get("https://api.nasdaq.com/api/screener/stocks", params=params_no, timeout=30)
```

```python
data = response.json()
ticker_count = len(data.get('data', {}).get('rows', []))
ticker_count
```

```python
data_no = response_no.json()
ticker_count_no = len(data_no.get('data', {}).get('rows', []))
ticker_count_no
```
